```python
from bs4 import BeautifulSoup as Soup
from urllib.request import urlopen as uReq
import requests
```

```python
url = "https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
resp = requests.get(url)
print(resp.text)
```

```python
s = Soup(resp.content,features='lxml')
s.select('._2cLu-l')[0].get_text() ##Product Name
```

```python
s.select('._1vC4OE')[0].get_text() ##Product Price
```

```python
s.select('.hGSR34')[0].get_text() ##Product Rating
```

```python
fname = "Flipkart Products Data.csv"
f = open(fname,'w')
```

```python
headers = "Product_Name, Price, Rating \n"
f.write(headers)
```

```python
for i in range(0,10):
    name = s.select('._2cLu-l')[i].get_text()
    price = s.select('._1vC4OE')[i].get_text()
    rating = s.select('.hGSR34')[i].get_text()
    print("Product Name:" + s.select('._2cLu-l')[0].get_text() )
    print("Price:" + s.select('._1vC4OE')[0].get_text())
    print("Rating:" + s.select('.hGSR34')[0].get_text())
    print(name.replace(","," |") + "," + price.replace("₹","Rs.").replace(",","") + "," +   rating + "\n")
    f.write(name.replace(","," |") + "," + price.replace("₹","Rs.").replace(",","") + "," +   rating + "\n")
f.close()    
```
